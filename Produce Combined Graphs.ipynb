```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
style = [":","--","-."]
```

## MAPE vs Prediction Horizon

```python
df = pd.read_csv("results/mape_vs_horizon.csv",index_col=0)
df
```

```python
df.plot(style=style)
plt.title("MAPE vs Prediction Horizon")
plt.ylabel("MAPE")
plt.xlabel("Prediction Horizon")
plt.ylim(0,0.5)
plt.legend()
```

```python
df = pd.read_csv("results/mape_vs_patients.csv",index_col=0)
df
```
