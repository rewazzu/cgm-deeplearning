```python
import pandas as pd
import openpyxl
from glob import glob
import os
```

```python
path = os.getcwd()
path += r"\Shanghai_T2DM"
csv_files = glob(os.path.join(path, "*.xls*"))
final=pd.DataFrame()
for file in csv_files:
    df = pd.read_excel(file).loc[:,["Date", "CGM (mg / dl)"]].dropna()
    df["ptid"] = file.split("\\")[-1][:4]
    df["Elapsed"] = df["Date"] - df["Date"].min() 
    row = df.pivot(index = "ptid", columns="Elapsed", values= "CGM (mg / dl)")
    final = pd.concat([final,row])
final = final.iloc[:,:480]
final = final.dropna()
final.to_csv("cgm_data.csv")
```
